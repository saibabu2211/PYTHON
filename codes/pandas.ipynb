```python
df["col"].apply(func) → apply func element-wise on a column.

df.apply(func, axis=0) → apply func to each column.

df.apply(func, axis=1) → apply func to each row.

groupby().apply(func) → custom group aggregations.
```

```python
df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})

print("map() on Series:")
print(df["A"].map(lambda x: x*10))

print("\napply() on DataFrame (row sum):")
print(df.apply(sum, axis=1))

print("\napplymap() on DataFrame (double every value):")
print(df.applymap(lambda x: x*2))
```

```python
df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT"],
    "Salary": [5000, 6000, 7000, 8000]
})

result = df.groupby("Department").agg(
    avg_salary=("Salary", "mean"),
    max_salary=("Salary", "max"),
    salary_range=("Salary", lambda x: x.max() - x.min())
)

print(result)
```

```python
df["Rank"] = df.groupby("Department")["Salary"].rank(method="dense", ascending=False)
df["Cumulative"] = df.groupby("Department")["Salary"].cumsum()
print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT"],
    "Year": [2023, 2024, 2023, 2024],
    "Salary": [5000, 6000, 7000, 8000]
})

pivot = pd.pivot_table(df, index=["Department", "Year"], values="Salary", aggfunc="sum")
print(pivot)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Day": range(1, 8),
    "Sales": [100, 200, 300, 400, 500, 600, 700]
})

# Rolling mean over last 3 days
df["Rolling_Avg"] = df["Sales"].rolling(window=3).mean()

print(df)
```

```python
df["Cumulative_Sum"] = df["Sales"].expanding().sum()
print(df)
```

```python
df["Rank"] = df["Sales"].rank(method="dense", ascending=False)
df["Lag1"] = df["Sales"].shift(1)   # previous row
df["Lead1"] = df["Sales"].shift(-1) # next row

print(df)
```

```python
data = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Finance"],
    "Employee": ["A", "B", "C", "D", "E"],
    "Salary": [5000, 6000, 7000, 8000, 9000]
})

pivot = pd.pivot_table(data,
                       index="Department",
                       values="Salary",
                       aggfunc="mean")

print(pivot)
```

```python
pivot = pd.pivot_table(data,
                       index="Department",
                       values="Salary",
                       aggfunc=["mean", "sum", "count"])

print(pivot)
```
